```python
import requests
import datetime
import json
from pprint import pprint
url="http://api.openweathermap.org/data/2.5/forecast?q=Rohini&appid=712ca061b3994e44162a359b1c23200a&units=metric"
req=requests.get(url)
data=req.json()
pprint((data))
temp=data["list"][3]['main']['temp']
humidity=data["list"][3]['main']['humidity']
weather=data["list"][3]['weather'][0]['main']
print("Time and Date now:",datetime.datetime.today())
print("Forecast for:",data["list"][2]["dt_txt"])
print("Temp:",temp,"°C","\nWeather:",weather,"\nHumidity:",humidity)
f= open("Weather_Python.txt","w+")
f.write("Time and Date now:"+str(datetime.datetime.today()))
f.write("\nForecast for:"+str(data["list"][3]["dt_txt"]))
f.write("\nTemp:"+str(temp)+"°C"+"\nWeather:"+str(weather)+"\nHumidity:"+str(humidity))
f.close()
```
